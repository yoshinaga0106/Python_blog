# PyStan

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
import sklearn
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
import pystan
from pystan import StanModel
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## データセット

```python
Boston = load_boston()
```

```python
X = Boston.data
y = Boston.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

線形回帰を行う
$$y = \alpha + X^T \beta  + \epsilon, \ \ \ \epsilon \sim N(0, \sigma)$$

## Stan側の設定

```python
dat = {'N': X_train.shape[0], 'M': X_train.shape[1], 'X': X_train, 'y': y_train}
```

```python
model = """
    data {
        int<lower=0> N;
        int<lower=0>M;
        matrix[N, M] X;
        vector[N] y;
    }
    parameters { 
        real beta_0;
        vector[M] beta;
        real<lower=0> sigma;
    }
    model { 
        for (i in 1:N)
            y[i] ~ normal(beta_0 + dot_product(X[i] , beta), sigma);
    }
"""
```

### コンパイル

```python
%time stm = StanModel(model_code=model)
```

```python
# コンパイルしたモデルを保存
import pickle
pickle.dump(stm, open('../model/model_linear_code.pkl', 'wb'))
```

```python
# stm = pickle.load(open('../model/model_linear_code.pkl', 'rb'))
```

## MCMC

```python
n_itr = 5000
n_warmup = 1000
chains = 2

# サンプリングの実行
%time fit = stm.sampling(data=dat, iter=n_itr, chains=chains, n_jobs=-1, warmup=n_warmup, algorithm="NUTS", verbose=False)
```

```python
pickle.dump(fit, open('../model/model_linear_boston.pkl', 'wb'))
```

```python
# fit = pickle.load(open('../model/model_linear_boston.pkl', 'rb'))
```

## 可視化

```python
fit.plot()
```

```python
# サンプル列を抽出
la  = fit.extract(permuted=True)
# パラメーター名
names = fit.model_pars 
# パラメーターの数
n_param = np.sum([1 if len(x) == 0 else x[0] for x in fit.par_dims])
```

```python
f, axes = plt.subplots(n_param, 2, figsize=(15, 4*n_param))
cnt = 0
for name in names:
    dat = la[name]
    if dat.ndim == 2:
        for j in range(dat.shape[1]):
            d = dat[:,j]
            sns.distplot(d, hist=False, ax=axes[cnt, 0])
            sns.tsplot(d,   alpha=0.8, lw=.5, ax=axes[cnt, 1])
            cnt += 1
    else:
        # Intercept
        sns.distplot(dat, hist=False, ax=axes[cnt, 0])
        sns.tsplot(dat,   alpha=0.8, lw=.5, ax=axes[cnt, 1])
        cnt += 1

name_list = []
for name in names:
    dat = la[name]
    if la[name].ndim == 2:
        for i in range(dat.shape[1]):
            name_list.append("{}{}".format(name,i+1))
    else:
        name_list.append(name)

for i in range(2):
    for j, t in enumerate(name_list):
        axes[j, i].set_title(t)
plt.show()
```

## 予実比較
各パラメータの平均値のリストを作成し、それをもとに予測値算出

```python
beta0 = la['beta_0'].mean()
beta = la['beta'].mean(axis = 0)
```

```python
y_pred_train = (X_train * beta).sum(axis = 1) + beta0
y_pred_test = (X_test * beta).sum(axis = 1) + beta0
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('MSE Train: %.2f, MSE Test: %.2f' % (mean_squared_error(y_train, y_pred_train), mean_squared_error(y_test, y_pred_test)))
print('R^2 Train: %.2f, R^2 Test: %.2f' % (r2_score(y_train, y_pred_train), r2_score(y_test, y_pred_test)))
```
